```python
isinGame = True

field = ["0","1","2","3","4","5","6","7","8"]
roundNumber = 1

def drawField():
    print()
    print("|-----------|")
    print("|",field[0], "|", field[1], "|", field[2], "|")
    print("|-----------|")
    print("|",field[3], "|", field[4], "|", field[5], "|")
    print("|-----------|")
    print("|",field[6], "|", field[7], "|", field[8], "|")
    print("|-----------|")    

def changeField(x):
    if roundNumber % 2 == 0:
        field[x] = "O"
    else:
        field[x] = "X"

def checkWin():
    line1 = field[0] + field[1] + field[2]
    line2 = field[3] + field[4] + field[5]
    line3 = field[6] + field[7] + field[8]
    lineV1 = field[0] + field[3] + field[6]
    lineV2 = field[1] + field[4] + field[7]
    lineV3 = field[2] + field[5] + field[8]
    lineP1 = field[0] + field[4] + field[8]
    lineP2 = field[2] + field[4] + field[6]

    WinConditionX = line1 == "XXX" or line2 == "XXX" or line3 == "XXX" or lineV1 == "XXX" or lineV2 == "XXX" or lineV3 == "XXX" or lineP1 == "XXX" or lineP2 == "XXX" 
    WinConditionO = line1 == "OOO" or line2 == "OOO" or line3 == "OOO" or lineV1 == "OOO" or lineV2 == "OOO" or lineV3 == "OOO" or lineP1 == "OOO" or lineP2 == "OOO" 
    return WinConditionX or WinConditionO

def swirchSign(roundNumber):
    if roundNumber % 2 == 0:
        return "O"
    else:
        return "X"

while isinGame:
    drawField()
    if checkWin():
        break
    print("Введите номер ячейки в которую хотите поставить {0}:".format(swirchSign(roundNumber)))
    x = int(input())
    changeField(x)
    roundNumber += 1
```
